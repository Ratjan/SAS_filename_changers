```python
import os
import re
import pandas as pd
import tkinter as tk
from tkinter.filedialog import askopenfilename
from tkinter.filedialog import askdirectory
from shutil import copyfile




```

```python
fn = askopenfilename()
print("user chose", fn)

df=pd.read_csv(fn, sep = ',')
df['scanid'] = df['Filename'].str.extract(r'(\d+)').astype(str) #extracts number from filename to scanid
```

```python
# Define the folder containing the CSV files
folder_path = askdirectory()
```

```python


# Function to check if the filename matches any scanid in the DataFrame
def check_Filename(filename, df):
    for scan_id in df['scanid'].values:
        if filename == scan_id:
            matching_rows = df[df['scanid'] == filename]
            if len(matching_rows) > 0:
                return matching_rows.iloc[0]  # Return the first matching row
    return None

# Iterate through all files in the folder
for file in os.listdir(folder_path):
    print(file)
    if file.endswith('.csv'):
        filename = os.path.basename(file)
        numstr = re.findall(r'\d+', filename)  # Find all numbers in the file name
        numstr = ''.join(numstr)
        result = check_Filename(numstr, df)  # Check if the numbers match any scanid

        if result is not None:
            new_filename = result['Sample Description'] + ".csv"
            new_filename = new_filename.replace(" ", "")  # Remove spaces from the new filename
            src = os.path.join(folder_path, file)
            dst = os.path.join(folder_path, new_filename)
            copyfile(src, dst)  # Make a copy with the new name
            print(f"Copied {file} to {new_filename}")
        else:
            print(f"No match found for {file}")
```
